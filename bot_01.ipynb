```python
import requests
import json
import pandas as pd
import time
```

```python
def download_all_markets_data():
    url = "https://api.bittrex.com/api/v1.1//public/getmarketsummaries"
    response = requests.get(url).text
    full_market_list = json.loads(response)['result']
    return pd.DataFrame(full_market_list)
```

```python
def filter_for_usable_markets(full_market_list_df):
    btc_markets = full_market_list_df['MarketName'].str.startswith('BTC')
    volume_ok = full_market_list_df['BaseVolume'] > 20
    buy_orders_ok = full_market_list_df['OpenBuyOrders'] > 150
    return full_market_list_df[btc_markets & volume_ok & buy_orders_ok]
```

```python
class Market:
    def __init__(self, name):
        self.name = name
        self.prices = []
        
    def load_price(self):
        url = 'https://api.bittrex.com/api/v1.1/public/getticker?market={}'.format(self.name)
        result = requests.get(url).text
        latest_price = json.loads(result)['result']
        self.prices.append(latest_price)
```

```python
amd = download_all_markets_data()
```

```python
ok_markets = filter_for_usable_markets(amd)
```

```python
markets = [Market(market_name) for market_name in ok_markets['MarketName']]
```

```python
for _ in range(100):
    for market in markets:
        market.load_price()
    time.sleep(5)
```

```python
for i,market in enumerate(markets):
    ceny = pd.DataFrame(market.prices)
    ceny['short_average'] = round(ceny['Last'].rolling(5).mean(), 6)
    ceny['long_average'] = round(ceny['Last'].rolling(20).mean(), 6)
    ceny['trend'] = ceny['short_average'] >= ceny['long_average']
    ceny = ceny.dropna()
    trend_ok = ceny['trend'] == True
    last_trend_ok = ceny['trend'].shift() == False
    ceny['enter'] = trend_ok & last_trend_ok
    if ceny['enter'].any():
        print(market.name, i)
```
